```python
#I will pass this method a number. It takes the number and returns a categorical classification or group
#This will allow me to bin the students to compare their performance based on if they are not full time (<12 hours),
#taking a typical course load (12-14 hours), taking a heavy course load (15-17 hours) or taking a particularly
#strenusous courseload (>=18 hours)
def classify_num_hours(num):
    if (num < 12):
        return '< 12'
    elif (num <= 14):
        return '12 - 14'
    elif (num <=17):
        return '15 - 17'
    else:
        return '>= 18'

#for every row, this code takes the value in pre_covid_courses['Total Course Hours Attempted'], passes it through the
#function above and stores the output in a new column called Hours Attempted Cat
pre_covid_courses['Hours_Attempted_Cat'] = pre_covid_courses['Total Course Hours Attempted'].apply(lambda x: classify_num_hours(x))
#this does the same as the line above but for the with_covid_courses dataframe
with_covid_courses['Hours_Attempted_Cat'] = with_covid_courses['Total Course Hours Attempted'].apply(lambda x: classify_num_hours(x))
```

```python
#This creates a new dataframe with only the student id's, terms and the category for the number of hours attempted 
#for the pre_covid_courses dataframe
info_pre = pd.DataFrame(pre_covid_courses[['Encrypt Id', 'Term Code', 'Hours_Attempted_Cat']])
#This does the same for the with covid courses dataframe
info_with = pd.DataFrame(with_covid_courses[['Encrypt Id', 'Term Code', 'Hours_Attempted_Cat']])
#These lines print out the first 5 lines of each dataframe
print(info_pre.head())
print(info_with.head())
```

```python
#This creates a dataframe with just a row for every student each term they took classes with their GPA for that term
#This one only considers the semesters prior to Spring 20
outcome_pre = pd.DataFrame(pre_covid[['Encrypt Id', 'Term', 'Semester GPA']])
#Does the same as above but only considers semesters including and after Spring 20
outcome_with = pd.DataFrame(with_covid[['Encrypt Id', 'Term', 'Semester GPA']])

#Prints the first 5 rows of each dataframe
print(outcome_pre.head())
print(outcome_with.head())
```

```python
#Creates a new column in info_pre with the student id followed by the term code, so for student 12345 in term 202010
#the value would be: 12345202010 this gives us a way to merge the outcome and info dataframes easily
info_pre['PK'] = info_pre['Encrypt Id'].astype(str) + info_pre['Term Code'].astype(str)
#Does the same for the second dataframe of info
info_with['PK'] = info_with['Encrypt Id'].astype(str) + info_with['Term Code'].astype(str)
```

```python
#This makes the matching column for each of the outcome dataframes
outcome_pre['PK'] = outcome_pre['Encrypt Id'].astype(str) + outcome_pre['Term'].astype(str)
outcome_with['PK'] = outcome_with['Encrypt Id'].astype(str) + outcome_with['Term'].astype(str)
```

```python
#Each of these commands merges the information about which category of course load the student was in each semester 
#with the semester gpa or outcome based on the student id-term code combo for both pre and with covid data
data_pre = info_pre.merge(outcome_pre, on = "PK")
data_with = info_with.merge(outcome_with, on = "PK")
```

```python
#prints the number of students in each course load category pre covid
data_pre['Hours_Attempted_Cat'].value_counts()
```

```python
#prints the number of students in each course load cateogry with covid
data_with['Hours_Attempted_Cat'].value_counts()
```

```python
#Drops any rows with any missing data in both dataframes
data_pre.dropna(inplace = True)
#the inplace = True parameter means that it drops the rows and puts the resulting dataframe back in the original object
data_with.dropna(inplace = True)
```

```python
#This groups by category and finds the average (mean) GPA in each category of courseload
grouped_pre = data_pre.groupby('Hours_Attempted_Cat').agg({'Semester GPA': np.mean})
grouped_pre
```

```python
#Repeats for the with covid data
grouped_with = data_with.groupby('Hours_Attempted_Cat').agg({'Semester GPA': np.mean})
grouped_with
```

```python
#Notice how many decimals are included above, the following commands round each column to just 2 places
grouped_pre['Semester GPA'] = grouped_pre['Semester GPA'].round(2)
grouped_with['Semester GPA'] = grouped_with['Semester GPA'].round(2)

#This renames the columns to have meaningful names so that we can merge the dataframes into one
grouped_pre.rename({'Semester GPA': 'GPA Pre-COVID'}, axis = 1, inplace = True)
grouped_with.rename({'Semester GPA': 'GPA With-COVID'}, axis = 1, inplace = True)

#This resets the index on each dataframe so that we can use the index values to merge into one dataframe
grouped_pre.reset_index(inplace = True)
grouped_with.reset_index(inplace = True)

#This merges the above dataframes into one that summarizes the results
grouped_hours = grouped_pre.merge(grouped_with, on = "Hours_Attempted_Cat")
grouped_hours
```

```python
#Notice how the above dataframe doesn't follow a logical order. We want the dataframe to go from smallest courseload
#to largest, so we make a list of the categories in the order we would like them
order = ['< 12', '12 - 14', '15 - 17', '>= 18']
#This makes a new column with the place we would like each category in a sorted dataframe
grouped_hours['sort_col'] = grouped_hours['Hours_Attempted_Cat'].apply(lambda x: order.index(x))
#This sorts based on the column we have created
grouped_hours.sort_values(by = 'sort_col', inplace = True)
grouped_hours
```

```python
#We can drop the extra column now that we don't need it any more
grouped_hours.drop('sort_col', axis = 1, inplace = True)
```

```python
ind = np.arange(4)  # the x locations for the groups
width = 0.35       # the width of the bars

fig = plt.figure() #creates a figure
ax = fig.add_subplot(111) #adds a plot to the figure
rects1 = ax.bar(ind, grouped_hours['GPA Pre-COVID'], width, color = 'blue') #on the plot we add one set of rectangles
#representing the bar for the GPA pre covid in each category

rects2 = ax.bar(ind+width, grouped_hours['GPA With-COVID'], width, color = 'orange') #does the same for the GPA with 
#covid for each category of course load

# add some features to the plot
ax.set_ylabel('Average GPA') #labels y axis
ax.set_title('Courseloads and the Pandemic') #gives plot a title
ax.set_xticks(ind + width / 2) #places the ticks on the x axis inbetween each rectange
ax.set_xticklabels(grouped_hours['Hours_Attempted_Cat']) #gives the xticks labels
ax.set_ylim(0, 4) #forces the y axis to go from 0 to 4

ax.legend( (rects1[0], rects2[0]), ('Pre-COVID', 'With-COVID') ) #gives labels for the legend

plt.show()
```

```python
# Keeps grades that are A, B, C, D and F and replaces all other codes with 'Other'
pre_covid_courses['Final Grade Category'] = pre_covid_courses['Final Grade'].apply(lambda x: x if x in ['A', 'B', 'C', 'D', 'F'] else 'Other')
with_covid_courses['Final Grade Category'] = with_covid_courses['Final Grade'].apply(lambda x: x if x in ['A', 'B', 'C', 'D', 'F'] else 'Other')
```

```python
#Finds the number of A's, number of B's, etc given prior to covid
grouped_pre = pre_covid_courses.groupby('Final Grade Category').agg({'Encrypt Id': 'count'})
#Turns the counts into percentages and rounds to 2 decimal places
grouped_pre['pre_percent'] = (grouped_pre['Encrypt Id'] / sum(grouped_pre['Encrypt Id']) * 100).round(2)
#Finds the number of A's, number of B's, etc given with covid
grouped_with = with_covid_courses.groupby('Final Grade Category').agg({'Encrypt Id': 'count'})
#Turns the counts into percentages and rounds to 2 decimal places
grouped_with['with_percent'] = (grouped_with['Encrypt Id'] / sum(grouped_with['Encrypt Id']) * 100).round(2)
#prints out calculations
print(grouped_pre)
print(grouped_with)
```

```python
#resets the index so we can merge the two dataframes above based on their indices
grouped_pre.reset_index(inplace = True)
grouped_with.reset_index(inplace = True)
#merge the dataframes
grouped = grouped_pre.merge(grouped_with, on = 'Final Grade Category')
#drop the counts of A's, B's, etcs for each
grouped.drop(['Encrypt Id_x', 'Encrypt Id_y'], axis = 1, inplace = True)
#Displays the resulting dataframe
grouped
```

```python
#plots the percentage for each grade given both pre covid (blue) and with covid (orange)
ind = np.arange(6)  # the x locations for the groups
width = 0.35       # the width of the bars

fig = plt.figure() #generates figure
ax = fig.add_subplot(111) #generates plot
rects1 = ax.bar(ind, grouped['pre_percent'], width, color = 'blue') #generates the set of bars for percentages pre-covid

rects2 = ax.bar(ind+width, grouped['with_percent'], width, color = 'orange') #generates the set of bars for percentatges with-covid
#and offsets them by the width of the first set

# add some extras
ax.set_ylabel('Percentage of Grades') #labels y axis
ax.set_title('Effect of the Pandemic on Grades') #adds title
ax.set_xticks(ind + width / 2) #adds location of x ticks
ax.set_xticklabels(grouped['Final Grade Category']) #adds labels for x ticks

ax.legend( (rects1[0], rects2[0]), ('Pre-COVID', 'With-COVID') ) #populates the legend

plt.show()
```

```python
#Returns quality points for each letter grade so that we can perform mathematical calculations
def weight(x):
    if (x == 'A'):
        return 4
    elif (x == 'B'):
        return 3
    elif (x == 'C'):
        return 2
    elif (x == 'D'):
        return 1
    elif (x == 'F'):
        return 0
    else:
        return None

#passes in the letter grade and fills a new column with the quality points associated with the letter grade
pre_covid_courses['Final Grade Weight'] = pre_covid_courses['Final Grade Category'].apply(lambda x: weight(x))
with_covid_courses['Final Grade Weight'] = with_covid_courses['Final Grade Category'].apply(lambda x: weight(x))
```
